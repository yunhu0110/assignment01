```python
import matplotlib.pyplot as plt
from scipy import signal
from skimage import io, color, exposure
import numpy as np
```

# Computing the magnitude of  gradient Function:

```python
def Magnitude(x,y) : 
    return np.sqrt(x**2+y**2)
```

# Computing the direction of gradient Function:

```python
def Distance(x,y) : # 두 벡터간의 거리를 계산한다
    return np.arctan(y/x)
```

```python
color_image = io.imread('cau.jpg')
gray_image  = color.rgb2gray(color_image)
kernel = np.array([[0,0,0],[0,1,0],[0,0,0]]) 
conv_image= signal.convolve2d(gray_image, ker, boundary='symm', mode='same') 

p1 = plt.subplot(2,2,1)
plt.imshow(color_image)
p1.set_title('color image')
plt.axis('off')

p2 = plt.subplot(2,2,2)
plt.imshow(gray_image, cmap='gray')
p2.set_title('Image of gray')
plt.axis('off')

p3 = plt.subplot(2,2,3)
plt.imshow(kernel, cmap='gray')
p3.set_title('kernel of Convoultion')
plt.axis('off')

p4 = plt.subplot(2,2,4)
plt.imshow(conv_image, cmap='gray')
p4.set_title('Result of Convolution')
plt.axis('off')

plt.show()
```
