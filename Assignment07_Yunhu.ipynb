```python
import numpy as np
import matplotlib.pyplot as plt

num     = 1001
std     = 5 

# x  : x-coordinate data
# y1 : (noisy) y-coordinate data
# y2 : (clean) y-coordinate data 

def fun(x):
 
	# f = np.sin(x) * (1 / (1 + np.exp(-x))) 
	f = np.abs(x) * np.sin(x)

	return f

n       = np.random.rand(num)
nn      = n - np.mean(n) #편차
x       = np.linspace(-10,10,num)
y1      = fun(x)
y2      = y1 + nn * std

plt.plot(x, y1, 'b.', x, y2, 'k.')
plt.show()
```

```python
def Approximation(X,Y):
    X = np.mat(X)
    Y = np.mat(Y).T
    xTx = X.T * X
    if(np.linalg.det(xTx) == 0.0):
        return
    w = xTx.I * (X.T * Y) 
    return w
```

```python
def Error(x, y):
    error = 0
    for i in range(0, len(x)):
        error += np.sqrt((float(x[i]) - float(y[i]))**2)
    print("MSE: ", error)        
    return np.sqrt(error)
```

```python
def func(n):
    D = []; dX = []; var = []; w = []; y = []; y_ = []
    for i in range(0, len(x)):
        dX = []
        for j in range(0, n+1):
                dX.append(x[i]**j)
        D.append(dX) 
    D = np.mat(D)
    w = Approximation(D, y2)
    print('weight:\n', w)
    for j in range(0, n+1):
        var = []
        var = (x**j)*(float(w[j]))
        y_.append(var)
    y_ = np.mat(y_)
    y_ = y_.T
    for i in range(0, len(y_)):
        sum_ = np.sum(y_[i])
        y.append(float(sum_))
    return y
```

```python
def training(n): 
    error = []; y = []
    y = func(n)
    error = Error(y2, y)
    Ylabel = n,' Dimension'
    visual(x, y, y2)
    return error
```

```python
def visual(x, y, y2):
    plt.plot(x, y, 'r', x, y2, 'b.')
    title ='Approximation'
    plt.title(title)
    plt.show()
```

```python
error = []
for n in range(0, 10):
    err = training(n)
    error.append(err)
```
